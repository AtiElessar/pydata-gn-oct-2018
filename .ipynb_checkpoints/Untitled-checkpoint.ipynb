# Objective: Find the sum of elements in this list

# C like indexing

```python
len_list = len(num_list)
sum_list = 0
for i in range(len_list):
    sum_list+= num_list[i]
```

### Let's make it a function

```python
def c_like_sum(l):
    len_list = len(num_list)
    sum_list = 0
    for i in range(len_list):
        sum_list+= num_list[i]
    return sum_list
```

```python
%timeit c_like_sum(num_list)
```

```python
def python_like_sum(l):
    sum_list = 0
    for x in l:
        sum_list+= x
    return sum_list
```

```python
%timeit python_like_sum(num_list)
```
